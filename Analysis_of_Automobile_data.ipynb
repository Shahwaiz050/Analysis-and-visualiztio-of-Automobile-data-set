```python
#Importing Libraries
import pandas as pd 
import numpy as np
```

```python
#header of the dataset
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#Importing the data set.
df = pd.read_csv('C:\\Users\\Shahwaiz Tahir\\Desktop\\Python Analysis\\auto.csv',names=headers)
df.head()
```

# Identify and handle the missing data

#Correct the cloumn name with normalized-losses

```python
#convert "?" to "NaN"
df.replace("?",np.nan,inplace=True)
df.head(5)
```

```python
#Check the null values
missing_data = df.isnull()
missing_data
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#find mean to replace the NaN values
mean1 = df["normalized-losses"].astype(float).mean()

print('average-normalized-losses:',mean1)
```

```python
#Replace NaN values by mean of the column in normalized-losses.
df["normalized-losses"].replace(np.nan,mean1,inplace=True)

#Check the column after replacing
df["normalized-losses"].head()
```

#Correct the column name with num-of-doors

```python
#To see which values are present in a particular column, we can use the ".value_counts()" method:

df["num-of-doors"].value_counts()
```

```python
#We can see that four doors are the most common type. We can also use the ".idxmax()" method to calculate the most common type automatically:

df["num-of-doors"].value_counts().idxmax()
```

```python
#The replacement procedure is very similar to what we have seen previously:

df["num-of-doors"].replace(np.nan,'four',inplace=True)
df["num-of-doors"].head()
```

#Corrext the column name with bore.

```python
#Look at the data

df['bore'].head()
```

```python
#Calculate the mean of bore column

mean_bore=df['bore'].astype('float').mean()
mean_bore
```

```python
#Replace the NaN value with mean in bore column
df['bore'].replace(np.nan,mean_bore,inplace=True)
df['bore'].isnull().sum()
```

# Correct the Stroke Column

```python
#Calculate the mean value of steoke column

mean_stroke = df['stroke'].astype('float').mean()
```

```python
#Replace the NaN value with mean of stroke
df['stroke'].replace(np.nan,mean_stroke,inplace=True)

#check the total null valules
df['stroke'].isnull().sum()
```

```python
df.head()
```

# Correct the horsepower column

```python
#Calculate the mean of horsepower column

mean_horse = df['horsepower'].astype('float').mean()

#Print the mean value
mean_horse
```

```python
#Replace the NaN value with mean
df['horsepower'].replace(np.nan,mean_horse,inplace=True)

#Check the total null value
df['horsepower'].isnull().sum()
```

# Correct the peak-rpm column 

```python
#Calculate the mean value of peak-rpm column
mean_peak = df['peak-rpm'].astype('float').mean()

#Replace the NaN value with mean values
df['peak-rpm'].replace(np.nan,mean_peak,inplace=True)

#Check the total null value
df['peak-rpm'].isnull().sum()
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

# Convert the data types

```python
df.dtypes
```

```python
#convert data types in correct format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df['city-mpg']

#Check the column name with city-L/100km
df[['city-L/100km']].head()
```

```python
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df['highway-mpg'] = 235/df['highway-mpg']

#Rename the column name highway-mpg with highway-L/100km 
df.rename(columns={'highway-mpg':'highway-L/100km'},inplace=True)

df.head()
```

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df[['length','width','height']].head()
```

# Binning data in Pandas

```python
#Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.
```

```python
#Convert data to correct format
df['horsepower'] = df['horsepower'].astype(int ,copy = True)
```

```python
#Let's plot the histogram of horsepower to see what the distribution of horsepower looks like.

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#build a bin array with a minimum value to a maximum value
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)

#set group names
group_names = ['Low', 'Medium', 'High']

#apply the function "cut" to determine what each value of df['horsepower'] belongs to.
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )

df['horsepower-binned'].head(10)
```

```python
df['horsepower-binned'].value_counts()
```

```python
#Let's plot the distribution of each bin
plt.pyplot.bar(group_names,df['horsepower-binned'].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('Horse Power Bins')
```
